# Show2D - Load TIFF

Use this notebook for TIFF files or TIFF folders.

```python
from pathlib import Path
from quantem.widget import Show2D

tiff_file = Path("data/stack.tiff")  # change this path

if tiff_file.exists():
    Show2D.from_tiff(tiff_file)
else:
    print(f"Set tiff_file to your TIFF path. Missing: {tiff_file}")
```

```python
tiff_folder = Path("data/tiff_stack")
if tiff_folder.exists():
    Show2D.from_tiff_folder(tiff_folder)
```
