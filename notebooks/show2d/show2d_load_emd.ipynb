# Show2D - Load EMD

Use this notebook for `.emd` files.

Note: EMD loading requires `h5py`.

```python
from pathlib import Path
from quantem.widget import Show2D

emd_file = Path("data/scan.emd")
dataset_path = "/data/signal"

if emd_file.exists():
    Show2D.from_emd(emd_file, dataset_path=dataset_path)
else:
    print(f"Set emd_file to your EMD path. Missing: {emd_file}")
```

```python
# Optional: reduce stack to one image
if emd_file.exists():
    Show2D.from_emd(emd_file, dataset_path=dataset_path, mode="index", index=0)
```
