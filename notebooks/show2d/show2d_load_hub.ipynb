# Show2D - Loading Hub
Choose one explicit loader:
1. PNG folder: `Show2D.from_png_folder(...)`
2. TIFF file/folder: `Show2D.from_tiff(...)` / `Show2D.from_tiff_folder(...)`
3. EMD file/folder: `Show2D.from_emd(...)` / `Show2D.from_emd_folder(...)`
For stacks, use optional reduction mode: `first`, `index`, `mean`, `max`, `sum`.

```python
from pathlib import Path
import quantem.widget
from quantem.widget import Show2D
print(f"quantem.widget {quantem.widget.__version__}")
```

```python
png_folder = Path("data/png_stack")
if png_folder.exists():
    Show2D.from_png_folder(png_folder)
```

```python
tiff_file = Path("data/stack.tiff")
if tiff_file.exists():
    Show2D.from_tiff(tiff_file)
```

```python
emd_file = Path("data/scan.emd")
dataset_path = "/data/signal"
if emd_file.exists():
    Show2D.from_emd(emd_file, dataset_path=dataset_path)
```
