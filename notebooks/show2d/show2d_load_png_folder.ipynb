# Show2D - Load PNG Folder
Use this notebook for a folder of PNG images.

```python
from pathlib import Path
import quantem.widget
from quantem.widget import Show2D
png_folder = Path("data/png_stack")  # change this path
if png_folder.exists():
    Show2D.from_png_folder(png_folder)
else:
    print(f"Set png_folder to your folder path. Missing: {png_folder}")
print(f"quantem.widget {quantem.widget.__version__}")
```

```python
# Optional: reduce stack to one image
if png_folder.exists():
    Show2D.from_png_folder(png_folder, mode="mean")
```
