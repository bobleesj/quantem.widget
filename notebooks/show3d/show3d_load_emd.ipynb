# Show3D - Load EMD
Use this notebook when your data is `.emd` (HDF-based).
Note: `.emd` loading requires `h5py` in your Python environment.

```python
from pathlib import Path
import quantem.widget
from quantem.widget import Show3D
emd_file = Path("data/scan.emd")  # change this path
dataset_path = "/data/signal"  # set to your dataset path in the EMD file
if emd_file.exists():
    w = Show3D.from_emd(emd_file, dataset_path=dataset_path)
    w
else:
    print(f"Set emd_file to your EMD file path. Missing: {emd_file}")
print(f"quantem.widget {quantem.widget.__version__}")
```

```python
# Mixed folder example: explicit type required
mixed_folder = Path("data/mixed_stack")
if mixed_folder.exists():
    Show3D.from_folder(mixed_folder, file_type="emd", dataset_path=dataset_path)
```
