# Show3D - Load TIFF
Use this notebook for single `.tif/.tiff` files or TIFF-only folders.

```python
from pathlib import Path
import quantem.widget
from quantem.widget import Show3D
tiff_file = Path("data/focal_series.tiff")  # change this path
if tiff_file.exists():
    w_file = Show3D.from_tiff(tiff_file)
    w_file
else:
    print(f"Set tiff_file to your TIFF file path. Missing: {tiff_file}")
print(f"quantem.widget {quantem.widget.__version__}")
```

```python
tiff_folder = Path("data/tiff_stack")  # optional: TIFF-only folder
if tiff_folder.exists():
    Show3D.from_tiff_folder(tiff_folder)
```
