# Show3D - Loading Hub

Pick one explicit loader path:

1. PNG folder -> `Show3D.from_png_folder(...)`
2. TIFF file/folder -> `Show3D.from_tiff(...)` / `Show3D.from_tiff_folder(...)`
3. EMD file/folder -> `Show3D.from_emd(...)` / `Show3D.from_emd_folder(...)` (optional `dataset_path`)

```python
from pathlib import Path
from quantem.widget import Show3D
```

```python
# PNG folder
png_folder = Path("data/png_stack")
if png_folder.exists():
    Show3D.from_png_folder(png_folder)
```

```python
# TIFF file
tiff_file = Path("data/focal_series.tiff")
if tiff_file.exists():
    Show3D.from_tiff(tiff_file)
```

```python
# EMD file
emd_file = Path("data/scan.emd")
dataset_path = "/data/signal"
if emd_file.exists():
    Show3D.from_emd(emd_file, dataset_path=dataset_path)
```
